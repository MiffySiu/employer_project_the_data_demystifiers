### UK datasets - salary bands

Data source and interactive plots: https://www.hesa.ac.uk/news/16-06-2022/sb263-higher-education-graduate-outcomes-statistics/salary<br>
Metadata: https://www.hesa.ac.uk/support/definitions/graduates<br>


Note:
- 22% of values are missing in the column "percent"
- cause salary is not mandatory, but self-reported
- catergory without missing values in 2017/18 and 2018/19:
    - Country of providers: ALL  (Region comparision is not possible as Wales got the most empties)
    - Domicile: ALL or ALL UK  (non-UK individuals are less willing to report)
    - Provider type: ALL
    - Level of qualification obtained: ALL or All postgraduate or All undergraduate
- use the same criterion in 2019/20:
    - missing values in male data
- There are two "Work population marker"
    - Paid employment is an activity -- employment 
    - Paid employment is most important activity  -- having activities more than employment

```python
# Check the data types
salary_1920.info()
```

```python
# remove the symbol %
salary_1920["Percent"] = salary_1920["Percent"].str.replace("%", "")

# check the changes
salary_1920.head()
```

```python
# change the data type of the column 'Percent'
salary_1920['Percent'] = salary_1920['Percent'].astype(float)

# Check the changes
salary_1920.head()
```

```python
# group the merged DataFrame by Salary band
salary_1920_employed = salary_1920[(salary_1920['Work population marker'] == "Paid employment is an activity") &
                                        (salary_1920['Country of provider'] == "All") &
                                        (salary_1920['Domicile'] == "All") &
                                        (salary_1920['Provider type'] == "All") &
                                        (salary_1920['Level of qualification obtained'] == "All") &
                                        (salary_1920['Mode of former study'] == "All")]
                                        
salary_1920_employed.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Use Seaborn
# The distribution of salary of all level in 2019-20
plt.subplots(figsize=(14,10))

sns.barplot(x="Salary band", y="Percent", estimator=sum, data=salary_1920_employed)

plt.xticks(rotation=90)

### The total percentage is 200% as this plot puts male and female together.
```

```python
# Use matplotlib
# The distribution of salary of all level in 2019-20
fig,ax = plt.subplots()
fig.set_size_inches(12, 8)

# create barplot
bars = ax.bar(salary_1920_employed['Salary band'], salary_1920_employed['Percent'], color="lightgrey")

plt.xticks(rotation=90)

plt.show()
```

```python
# The distribution of salary of all level in 2019-20, by gender
plt.subplots(figsize=(14,10))

sns.barplot(x="Salary band", y="Percent", 
            hue="Personal characteristic category", 
            estimator=sum, data=salary_1920_employed)

plt.xticks(rotation=90)
```
