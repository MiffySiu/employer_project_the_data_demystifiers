### UK datasets - salary bands

Data source and interactive plots: https://www.hesa.ac.uk/news/16-06-2022/sb263-higher-education-graduate-outcomes-statistics/salary<br>
Metadata: https://www.hesa.ac.uk/support/definitions/graduates<br>


Note:
- 22% of values are missing in the column "percent"
- cause salary is not mandatory, but self-reported
- catergory without missing values in 2017/18 and 2018/19:
    - Country of providers: ALL  (Region comparision is not possible as Wales got the most empties)
    - Domicile: ALL or ALL UK  (non-UK individuals are less willing to report)
    - Provider type: ALL
    - Level of qualification obtained: ALL or All postgraduate or All undergraduate
- use the same criterion in 2019/20:
    - missing values in male data
- There are two "Work population marker"
    - Paid employment is an activity -- employment 
    - Paid employment is most important activity  -- having activities more than employment

```python
# remove the symbol %
salary_1920["Percent"] = salary_1920["Percent"].str.replace("%", "")

# check the changes
salary_1920.head()
```

```python
# remove the symbol %
salary_1718["Percent"] = salary_1718["Percent"].str.replace("%", "")

# check the changes
salary_1718.head()
```

```python
# change the data type of the column 'Percent'
salary_1718['Percent'] = salary_1718['Percent'].astype(float)

# Check the changes
salary_1718.head()
```

```python
# Check the change of data type
salary_1718.info()
```

```python
# group the merged DataFrame by Salary band
salary_1718_employed = salary_1718[(salary_1718['Work population marker'] == "Paid employment is an activity") &
                                        (salary_1718['Country of provider'] == "All") &
                                        (salary_1718['Domicile'] == "All") &
                                        (salary_1718['Provider type'] == "All") &
                                        (salary_1718['Level of qualification obtained'] == "All") &
                                        (salary_1718['Mode of former study'] == "All")].reset_index().copy()

# Check the results                                        
salary_1718_employed.head()
```

```python
# Check for missing values
print(salary_1718_employed.isnull().sum())
```

```python
# Merge the subsets
salary = pd.concat([salary_1920_employed, salary_1819_employed, salary_1718_employed])

# Check the total number of rows
salary.info()
```

```python
# Drop the unnecessary columns
salary.drop(columns=['index',
                     'Country of provider',
                     'Domicile',
                     'Provider type',
                     'Level of qualification obtained',
                     'Mode of former study',
                     'Work population marker'],
             inplace=True)

# Check the changes
salary.head()
```

```python
salary.columns = ['Gender', 'Salary band', 'Year', 'Percent']

salary.head()
```

```python
salary_female = salary[salary['Gender'] == "Female"].reset_index().copy()

salary_female.head()
```

```python
salary_male = salary[salary['Gender'] == "Male"].reset_index().copy()

salary_male.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# The distribution of salary of all level
plt.subplots(figsize=(14,10))

sns.barplot(x="Salary band", y="Percent", 
            hue="Year", 
            estimator=sum, data=salary_female)

plt.xticks(rotation=90)
```

```python
# The distribution of salary of all level
plt.subplots(figsize=(14,10))

sns.barplot(x="Salary band", y="Percent", 
            hue="Year", 
            estimator=sum, data=salary_male)

plt.xticks(rotation=90)
```

```python
# The distribution of salary of all level in 2019-20, by gender
plt.subplots(figsize=(14,10))

sns.barplot(x="Salary band", y="Percent", 
            hue="Personal characteristic category", 
            estimator=sum, data=salary_1920_employed)

plt.xticks(rotation=90)
```
